```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from konlpy.tag import Okt
from collections import Counter

# 1. 텍스트 데이터 합치기
# df['dream_content']에 있는 모든 텍스트를 하나의 긴 문자열로 만듭니다.
# 혹시 데이터가 문자열이 아닐 경우를 대비해 .astype(str)를 붙입니다.
text = " ".join(df['dream_content'].astype(str).tolist())

# 2. 형태소 분석기(Okt)로 명사만 추출하기
okt = Okt()
print("형태소 분석 중... 잠시만 기다려주세요.")
nouns = okt.nouns(text) # 명사만 추출

# 3. 불용어(제외할 단어) 처리 및 한 글자 단어 제거
# 꿈 일기 특성상 '꿈', '나', '것' 같은 단어는 너무 많이 나오므로 제외합니다.
stop_words = ['나', '너', '그', '것', '수', '등', '이', '꿈', '속', '안', '위', '때문']
nouns = [n for n in nouns if n not in stop_words and len(n) > 1]

# 4. 단어 빈도수 계산
counts = Counter(nouns)
print(f"추출된 주요 단어 Top 5: {counts.most_common(5)}") # 상위 5개 확인

# 5. 워드클라우드 설정 (Mac 한글 폰트 경로 필수)
wc = WordCloud(
    font_path='/System/Library/Fonts/AppleSDGothicNeo.ttc', # Mac 기본 폰트
    background_color='white',      # 배경색 (black, white 등)
    width=1000, 
    height=1000,
    max_words=100,                 # 최대 표현할 단어 수
    max_font_size=300,             # 가장 빈도 높은 단어의 크기
    colormap='viridis'             # 색상 테마 (prism, cool, viridis, plasma 등)
)

# 6. 워드클라우드 생성 및 시각화
cloud = wc.generate_from_frequencies(counts)

plt.figure(figsize=(10, 10))
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off') # x, y축 눈금 제거
plt.show()
```

```python
# emotion_score 평균 계산
import json
import glob

# 데이터 불러오기
file_pattern = './test_dataset/december/*.json'
file_list = glob.glob(file_pattern)

emotion_scores = []

for file_path in file_list:
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)
            if 'emotion_score' in data:
                emotion_scores.append(data['emotion_score'])
    except Exception as e:
        print(f"Error reading {file_path}: {e}")

# 평균 계산
if emotion_scores:
    average_score = sum(emotion_scores) / len(emotion_scores)
    print(f"\n=== 이번 달 감정 점수 분석 ===")
    print(f"총 데이터 수: {len(emotion_scores)}개")
    print(f"평균 감정 점수: {average_score:.2f}점")
    print(f"최고 점수: {max(emotion_scores)}점")
    print(f"최저 점수: {min(emotion_scores)}점")
    
    # 점수별 분포
    score_distribution = {i: emotion_scores.count(i) for i in range(1, 6)}
    print(f"\n점수별 분포:")
    for score, count in sorted(score_distribution.items()):
        percentage = (count / len(emotion_scores)) * 100
        print(f"  {score}점: {count}일 ({percentage:.1f}%)")
else:
    print("감정 점수 데이터가 없습니다.")
```

# 알파버전 : 해당 월에서 특정 감정 점수 글만 샘플로 가져오기

```python
import json
import glob
from collections import defaultdict

def extract_representative_summaries(data_folder='./test_dataset/december'):
    """감정 점수별로 대표 today_fortune_summary 추출"""
    
    # 1. 데이터 로드
    all_data = []
    for file_path in sorted(glob.glob(f'{data_folder}/*.json')):
        with open(file_path, 'r', encoding='utf-8') as f:
            all_data.append(json.load(f))
    
    # 2. 감정별로 그룹화
    summaries_by_emotion = defaultdict(list)
    for data in all_data:
        summaries_by_emotion[data['emotion_score']].append({
            'date': data['date'],
            'summary': data['today_fortune_summary'],
            'dream_interpretation': data['dream_interpretation']
        })
    
    # 3. 각 감정 카테고리에서 대표 선택
    representative = []
    
    # 극단적 감정 우선 (1점, 2점, 5점)과 중간값 포함
    priority_emotions = [5, 1, 2, 4, 3]
    
    for emotion in priority_emotions:
        if summaries_by_emotion[emotion]:
            # 각 감정에서 2개씩 (총 10개 정도면 충분)
            count = 4 if emotion in [5, 1, 2] else 1
            for item in summaries_by_emotion[emotion][:count]:
                representative.append({
                    'emotion': emotion,
                    **item
                })
    
    # 4. 통계
    emotion_scores = [d['emotion_score'] for d in all_data]
    stats = {
        'total_days': len(all_data),
        'avg_emotion': round(sum(emotion_scores) / len(emotion_scores), 2),
        'emotion_distribution': {i: emotion_scores.count(i) for i in range(1, 6)},
        'period': all_data[0]['date'][:7],
        'user_name': all_data[0]['user_name'],
        'birth_date': all_data[0]['birth_date']
    }
    
    return representative, stats, all_data

def create_monthly_analysis_prompt(representative, stats):
    """LLM에게 보낼 최적화된 프롬프트 생성"""
    
    emotion_labels = {
        1: "매우 부정적 (슬픔, 우울, 무기력)",
        2: "부정적 (불안, 피곤)",
        3: "중립 (평범함, 무감정)",
        4: "긍정적 (기대, 편안함)",
        5: "매우 긍정적 (행복, 기쁨, 희열)"
    }
    
    prompt = f"""당신은 꿈과 심리 분석 전문가입니다. {stats['user_name']}님({stats['birth_date']}생)의 {stats['period']} 한 달간 꿈 일기를 분석하여 감동적이고 서정적인 월간 리포트를 작성해주세요.

## 📊 한 달 개요
- 총 {stats['total_days']}일 꿈 기록
- 평균 감정 점수: {stats['avg_emotion']}/5.0
- 감정 분포:
"""
    
    for score in range(5, 0, -1):
        count = stats['emotion_distribution'][score]
        percentage = (count / stats['total_days']) * 100
        prompt += f"  · {emotion_labels[score]}: {count}일 ({percentage:.1f}%)\n"
    
    prompt += f"\n## 🌙 대표적인 날들의 운세 분석\n\n"
    
    for idx, item in enumerate(representative, 1):
        prompt += f"### {idx}. [{item['date']}] 감정 점수 {item['emotion']}점\n"
        prompt += f"**꿈 해몽**: {item['dream_interpretation']}\n\n"
        prompt += f"**상세 분석**:\n{item['summary']}\n\n"
        prompt += "---\n\n"
    
    prompt += """## ✍️ 월간 리포트 작성 요청

위의 대표적인 날들을 중심으로, 한 달 전체의 감정 흐름과 꿈의 패턴을 분석하여 다음 구조로 작성해주세요:

### 0. 인삿말. 해당 월의 날씨 특징을 가볍게 은유하여 사용자에게 감동을 주세요.

### 1. 이번 달의 감정 여정 (300-400자)
- 한 달간 감정의 기복과 흐름을 서정적으로 표현
- 대표적인 날들을 예시로 들며 이야기체로 서술
- "○월 초반에는...", "중순 무렵...", "말에는..." 식으로 시간의 흐름 표현

### 2. 꿈이 전하는 메시지 (400-500자)
- 반복되는 꿈의 주제나 상징 분석
- 감정 패턴의 의미 해석
- 무의식이 말하고자 하는 것
- 구체적인 날짜와 꿈을 언급하며 설득력 있게

### 3. 심리적 통찰과 성장 (300-400자)
- 이번 달 경험한 내면의 변화
- 꿈을 통해 발견한 자신의 모습
- 성장의 증거들

### 4. 다음 달을 위한 따뜻한 조언 (200-250자)
- 구체적이면서도 따뜻한 격려
- 이번 달 배운 것을 바탕으로 한 실천 가능한 제안
- 희망적이고 감동적인 마무리

**작성 가이드**:
- 통계 숫자는 자연스럽게 녹여서 표현
- 사용자 이름({stats['user_name']}님)을 자연스럽게 사용
- 감동적이고 개인화된 느낌으로
- 심리학적 깊이와 문학적 아름다움을 동시에
- 각 섹션은 명확히 구분하되 자연스럽게 연결"""

    return prompt


# 실행
representative, stats, all_data = extract_representative_summaries()
prompt = create_monthly_analysis_prompt(representative, stats)

print(prompt)
print(f"\n\n=== 메타정보 ===")
print(f"선택된 대표 일수: {len(representative)}개")
print(f"전체 데이터: {len(all_data)}개")
print(f"압축률: {len(representative)/len(all_data)*100:.1f}%")
```

```python
async def generate_monthly_report_streaming():
    """월간 리포트 생성 (스트리밍 버전)"""
    
    print("=" * 60)
    print("🌙 월간 꿈 분석 리포트 생성 시작")
    print("=" * 60)
    
    # 1. 데이터 추출 및 프롬프트 생성
    print("\n[1/3] 데이터 추출 및 프롬프트 생성 중...")
    representative, stats, all_data = extract_representative_summaries()
    prompt = create_monthly_analysis_prompt(representative, stats)
    
    print(f"✅ 완료")
    print(f"   - 전체 데이터: {len(all_data)}일")
    print(f"   - 대표 선택: {len(representative)}일")
    print(f"   - 압축률: {len(representative)/len(all_data)*100:.1f}%")
    print(f"   - 프롬프트 길이: {len(prompt):,}자")
    
    # 2. API 호출 (스트리밍)
    print("\n[2/3] GMS API 호출 중 (실시간 스트리밍)...")
    print("=" * 60)
    print()
    
    monthly_report = ""
    
    try:
        stream = await client.chat.completions.create(
            model='gpt-5-mini',
            messages=[
                {
                    "role": "system", 
                    "content": "당신은 꿈과 심리 분석 전문가입니다. 감동적이고 서정적인 글을 작성합니다."
                },
                {
                    "role": "user", 
                    "content": prompt
                }
            ],
            stream=True,
        )
        
        async for chunk in stream:
            if chunk.choices[0].delta.content is not None:
                content = chunk.choices[0].delta.content
                monthly_report += content
                print(content, end="", flush=True)
        
        print("\n")
        print("=" * 60)
        print("\n[3/3] 생성 완료!")
        
        # 3. 결과 저장
        output = {
            'period': stats['period'],
            'user_name': stats['user_name'],
            'statistics': stats,
            'report': monthly_report,
            'metadata': {
                'total_days': len(all_data),
                'representative_days': len(representative),
                'prompt_length': len(prompt)
            }
        }
        
        # JSON 파일로 저장
        output_file = f"monthly_report_{stats['period']}.json"
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(output, f, ensure_ascii=False, indent=2)
        
        print(f"\n💾 결과가 '{output_file}'에 저장되었습니다.")
        
        return output
        
    except Exception as e:
        print(f"\n❌ 오류 발생: {str(e)}")
        return None
```

```python
def create_compact_monthly_prompt(data_folder='./test_dataset/december'):
    """31일 전체를 압축된 형태로 프롬프트에 포함"""
    
    # 데이터 로드
    all_data = []
    for file_path in sorted(glob.glob(f'{data_folder}/*.json')):
        with open(file_path, 'r', encoding='utf-8') as f:
            all_data.append(json.load(f))
    
    # 통계
    emotion_scores = [d['emotion_score'] for d in all_data]
    stats = {
        'total_days': len(all_data),
        'avg_emotion': round(sum(emotion_scores) / len(emotion_scores), 2),
        'emotion_distribution': {i: emotion_scores.count(i) for i in range(1, 6)},
        'period': all_data[0]['date'][:7],
        'user_name': all_data[0]['user_name'],
        'birth_date': all_data[0]['birth_date']
    }
    
    emotion_labels = {
        1: "매우 부정적", 2: "부정적", 3: "중립", 4: "긍정적", 5: "매우 긍정적"
    }
    
    prompt = f"""당신은 꿈 분석 전문가입니다. {stats['user_name']}님({stats['birth_date']}생)의 {stats['period']} 한 달 꿈 일기를 분석해주세요.

## 통계
- 총 {stats['total_days']}일 기록
- 평균 감정: {stats['avg_emotion']}/5.0
- 분포: 😊{stats['emotion_distribution'][5]}일 🙂{stats['emotion_distribution'][4]}일 😐{stats['emotion_distribution'][3]}일 😟{stats['emotion_distribution'][2]}일 😢{stats['emotion_distribution'][1]}일

## 날짜별 운세 분석 (전체 31일)

"""
    
    # 31일 전체를 간결하게 포함
    for data in all_data:
        # today_fortune_summary의 핵심만 추출 (첫 2문장 또는 200자)
        summary = data['today_fortune_summary']
        # 마침표나 느낌표 기준으로 앞 2문장만
        sentences = summary.replace('다. ', '다.|').replace('니다. ', '니다.|').split('|')
        compact_summary = sentences[0] if sentences else summary[:200]
        
        prompt += f"**[{data['date']}] 감정{data['emotion_score']}점** - {data['dream_interpretation']}\n"
        prompt += f"{compact_summary}\n\n"
    
    prompt += f"""
## 작성 요청

위 31일 전체의 흐름을 분석하여 감동적인 월간 리포트를 작성해주세요:

### 1. 이번 달의 감정 여정 (300-400자)
한 달간의 감정 기복과 흐름을 서정적으로. 초반/중반/후반의 변화를 포함.

### 2. 꿈이 전하는 메시지 (400-500자)
반복되는 주제, 패턴, 무의식의 메시지를 구체적 날짜와 함께.

### 3. 심리적 통찰과 성장 (300-400자)
내면의 변화와 성장의 증거들.

### 4. 다음 달을 위한 조언 (200-250자)
따뜻하고 실천 가능한 격려.

**스타일**: 감동적, 개인화, 서정적. {stats['user_name']}님을 자연스럽게 호칭.

각 섹션 구분은 명확하되 흐름은 자연스럽게하고 사용자에겐 목차를 보여주지 마세요.
"""

    return prompt, stats
```

```python
def create_keyword_based_prompt(data_folder='./test_dataset/december'):
    """각 날의 핵심 키워드만 추출해서 전체 포함"""
    
    all_data = []
    for file_path in sorted(glob.glob(f'{data_folder}/*.json')):
        with open(file_path, 'r', encoding='utf-8') as f:
            all_data.append(json.load(f))
    
    emotion_scores = [d['emotion_score'] for d in all_data]
    stats = {
        'total_days': len(all_data),
        'avg_emotion': round(sum(emotion_scores) / len(emotion_scores), 2),
        'emotion_distribution': {i: emotion_scores.count(i) for i in range(1, 6)},
        'period': all_data[0]['date'][:7],
        'user_name': all_data[0]['user_name'],
        'birth_date': all_data[0]['birth_date']
    }
    
    prompt = f"""당신은 꿈 분석 전문가입니다. {stats['user_name']}님의 {stats['period']} 한 달 꿈 일기 분석.

## 통계
총 {stats['total_days']}일 | 평균 감정 {stats['avg_emotion']}/5.0
분포: 매우긍정{stats['emotion_distribution'][5]} 긍정{stats['emotion_distribution'][4]} 중립{stats['emotion_distribution'][3]} 부정{stats['emotion_distribution'][2]} 매우부정{stats['emotion_distribution'][1]}

## 전체 31일 꿈 기록

"""
    
    # 날짜별로 한 줄씩
    for data in all_data:
        # 해몽만 사용 (가장 압축된 형태)
        prompt += f"{data['date']} (감정{data['emotion_score']}): {data['dream_interpretation']}\n"
    
    prompt += f"""

## 상세 분석 (중요한 날들)

"""
    
    # 극단적인 날들만 상세 설명 추가
    extreme_days = sorted(all_data, key=lambda x: abs(x['emotion_score'] - 3), reverse=True)[:8]
    
    for data in extreme_days:
        prompt += f"**[{data['date']}] 감정 {data['emotion_score']}점**\n"
        prompt += f"{data['today_fortune_summary'][:300]}...\n\n"
    
    prompt += """
## 월간 리포트 작성

위 전체 31일의 흐름과 주요 날들을 바탕으로:

1. **이번 달의 감정 여정** (300-400자): 초중후반 흐름을 서정적으로
2. **꿈이 전하는 메시지** (400-500자): 패턴과 의미
3. **심리적 통찰과 성장** (300-400자): 내면의 변화
4. **다음 달 조언** (200-250자): 따뜻한 격려

감동적이고 개인화된 느낌으로 작성.
각 섹션 구분은 명확하되 흐름은 자연스럽게하고 사용자에겐 목차를 보여주지 마세요.
"""

    return prompt, stats
```

```python
def create_full_detail_prompt(data_folder='./test_dataset/december'):
    """31일 전체 상세 내용 모두 포함"""
    
    all_data = []
    for file_path in sorted(glob.glob(f'{data_folder}/*.json')):
        with open(file_path, 'r', encoding='utf-8') as f:
            all_data.append(json.load(f))
    
    emotion_scores = [d['emotion_score'] for d in all_data]
    stats = {
        'total_days': len(all_data),
        'avg_emotion': round(sum(emotion_scores) / len(emotion_scores), 2),
        'emotion_distribution': {i: emotion_scores.count(i) for i in range(1, 6)},
        'period': all_data[0]['date'][:7],
        'user_name': all_data[0]['user_name'],
        'birth_date': all_data[0]['birth_date']
    }
    
    prompt = f"""# {stats['user_name']}님의 {stats['period']} 월간 꿈 분석

## 기본 정보
- 이름: {stats['user_name']}
- 생년월일: {stats['birth_date']}
- 분석 기간: {stats['period']} (총 {stats['total_days']}일)
- 평균 감정 점수: {stats['avg_emotion']}/5.0

## 감정 분포
- 매우 긍정적 (5점): {stats['emotion_distribution'][5]}일
- 긍정적 (4점): {stats['emotion_distribution'][4]}일
- 중립 (3점): {stats['emotion_distribution'][3]}일
- 부정적 (2점): {stats['emotion_distribution'][2]}일
- 매우 부정적 (1점): {stats['emotion_distribution'][1]}일

## 날짜별 상세 꿈 운세 분석

"""
    
    # 31일 전체를 상세하게 모두 포함
    for idx, data in enumerate(all_data, 1):
        prompt += f"### Day {idx}. {data['date']} (감정 점수: {data['emotion_score']}/5)\n\n"
        prompt += f"**꿈 해몽**: {data['dream_interpretation']}\n\n"
        prompt += f"**오늘의 운세 분석**:\n{data['today_fortune_summary']}\n\n"
        prompt += "---\n\n"
    
    prompt += """
## 월간 리포트 작성 요청

위 31일 전체의 꿈과 운세를 종합 분석하여, 감동적이고 서정적인 월간 리포트를 작성해주세요.

### 구성
1. **인사말** (50자): 해당 월의 날씨/계절 특징을 은유적으로
2. **이번 달의 감정 여정** (300-400자): 초중후반의 감정 흐름을 이야기체로
3. **꿈이 전하는 메시지** (400-500자): 반복 주제, 패턴, 구체적 날짜 언급
4. **심리적 통찰과 성장** (300-400자): 내면의 변화와 성장
5. **다음 달을 위한 조언** (200-250자): 따뜻하고 실천 가능한 격려

### 작성 가이드
- 사용자 이름을 자연스럽게 사용
- 감동적이고 개인화된 느낌
- 심리학적 깊이 + 문학적 아름다움
- 통계 수치는 자연스럽게 녹여서
- 각 섹션 구분은 명확하되 흐름은 자연스럽게하고 사용자에겐 목차를 보여주지 마세요

"""

    return prompt, stats
```

```python
async def generate_monthly_report_with_prompt(prompt, stats):
    """주어진 프롬프트로 월간 리포트 생성 (스트리밍)"""
    
    print("=" * 60)
    print("🌙 월간 꿈 분석 리포트 생성 시작")
    print("=" * 60)
    
    print(f"\n프롬프트 정보:")
    print(f"   - 길이: {len(prompt):,}자")
    print(f"   - 예상 토큰: ~{len(prompt)//4:,} tokens")
    print(f"   - 기간: {stats['period']}")
    print(f"   - 사용자: {stats['user_name']}님")
    
    print("\n[API 호출 중] 실시간 스트리밍...")
    print("=" * 60)
    print()
    
    monthly_report = ""
    
    try:
        stream = await client.chat.completions.create(
            model='gpt-4.1',
            messages=[
                {
                    "role": "system", 
                    "content": "당신은 꿈과 심리 분석 전문가입니다. 감동적이고 서정적인 글을 작성합니다."
                },
                {
                    "role": "user", 
                    "content": prompt
                }
            ],
            max_tokens=2500,
            temperature=0.8,
            stream=True,
        )
        
        async for chunk in stream:
            if chunk.choices[0].delta.content is not None:
                content = chunk.choices[0].delta.content
                monthly_report += content
                print(content, end="", flush=True)
        
        print("\n")
        print("=" * 60)
        print("\n✅ 생성 완료!")
        
        # 결과 저장
        output = {
            'period': stats['period'],
            'user_name': stats['user_name'],
            'statistics': stats,
            'report': monthly_report,
            'metadata': {
                'total_days': stats['total_days'],
                'prompt_length': len(prompt),
                'avg_emotion': stats['avg_emotion']
            }
        }
        
        # JSON 파일로 저장
        output_file = f"monthly_report_{stats['period']}.json"
        with open(output_file, 'w', encoding='utf-8') as f:
            json.dump(output, f, ensure_ascii=False, indent=2)
        
        print(f"\n💾 결과가 '{output_file}'에 저장되었습니다.")
        
        return output
        
    except Exception as e:
        print(f"\n❌ 오류 발생: {str(e)}")
        import traceback
        traceback.print_exc()
        return None
```
